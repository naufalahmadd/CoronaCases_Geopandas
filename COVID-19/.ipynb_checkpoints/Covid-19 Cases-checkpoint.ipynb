```python
import requests
from bs4 import BeautifulSoup

uri = 'https://www.worldometers.info/coronavirus/'
headers = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

r = requests.get(uri, headers=headers)
soup = BeautifulSoup(r.text, 'lxml')

table = soup.findAll('table')[0]

data = []
columns = [col.getText(separator=' ', strip=True).replace('\xa0', '') 
           for col in table.findAll('thead')[0].findAll('th')]
table_body = table.findAll('tbody')[0]
rows = table_body.findAll('tr')

for row in rows:
    cols = [col.text.strip() for col in row.findAll('td')]
    data.append(cols)

data_df = pd.DataFrame(data, columns=columns)
data_df.head()
```
